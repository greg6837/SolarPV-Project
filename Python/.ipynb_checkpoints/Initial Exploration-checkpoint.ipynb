```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Import CSI Dataset

```python
df_csi = pd.read_csv('../Data/ca-solar-2018-04-29.csv')
df_CAinc = pd.read_csv('../Data/2016_5YR_Cal_inc.csv')
```

```python
df_csi.shape
```

```python
df_csi.head()
```

```python
df_csi.isnull().sum()
```

```python
df_csi = df_csi.dropna(axis = 1, how = 'all')
```

```python
plt.figure(figsize = (10,10))
df_csi['Host Customer Physical Address County'].value_counts().plot(kind = 'bar')
plt.title('Customer count in each County in California')
plt.xlabel('County')
plt.ylabel('Count')
```

```python
plt.figure(figsize = (8,6))
df_csi['Host Customer Sector'].value_counts().plot(kind = 'bar')
plt.title('Count of Customer Types in California')
plt.xlabel('Customer Types')
plt.ylabel('Count')
```
